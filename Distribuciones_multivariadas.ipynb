## Distribuciones multivariadas

Let $X$ and $Y$ be two continuous random variables with joint probability density function
$$
f(x, y)=\frac{12}{5} x y(1+y) \quad \text { for } 0 \leq x \leq 1 \text { and } 0 \leq y \leq 1
$$
and $f(x, y)=0$ otherwise.


---



Find the probability $\mathrm{P}\left(\frac{1}{4} \leq X \leq \frac{1}{2}, \frac{1}{3} \leq Y \leq \frac{2}{3}\right)$

\begin{equation*}
\begin{split}
\mathrm{P}\left(\frac{1}{4} \leq X \leq \frac{1}{2}, \frac{1}{3} \leq Y \leq \frac{2}{3}\right)&= \int_{1/4}^{1/2} \int_{1/3}^{2/3} \frac{12}{5}xy(1+y) dydx\\
&= \int_{1/4}^{1/2} \frac{12}{5}x \left( \int_{1/3}^{2/3} y(1+y) dy\right)dx\\
&= \int_{1/4}^{1/2}  \frac{82}{135}x \ dx\\
&= \frac{41}{720}
\end{split}
\end{equation*}

Determine the joint distribution function of $X$ and $Y$ for $a$ and $b$ between 0 and 1

\begin{equation*}
\begin{split}
F(a,b) &= P(X \leq a,Y \leq b)\\
&= \int_0^a \int_0^b f(x,y) dy dx\\
&= \int_0^a \int_0^b \frac{12}{5}xy(1+y) dy dx\\
&= \int_0^a \frac{12}{5}x \int_0^b y(1+y) dy dx\\
&= \int_0^a \frac{12}{5}x \left[ \frac{b^2}{2} + \frac{b^3}{3}\right] dx\\
&= \left[ \frac{b^2}{2} + \frac{b^3}{3}\right] \int_0^a \frac{12}{5}x  dx\\
&= \frac{12}{5}\left[ \frac{b^2}{2} + \frac{b^3}{3}\right]
\left[ \frac{a^2}{2}\right]\\
&= \frac{3}{5}a^2b^2 + \frac{2}{5}a^2b^3
\end{split}
\end{equation*}


Use your answer from b to find $F_{X}(a)$ for $a$ between 0 and 1

\begin{equation*}
\begin{split}
F_{X}(a) &= P(X\leq a)\\
&= \lim_{b \to \infty} F(a,b)\\
&= \lim_{b \to 1} \frac{3}{5}a^2b^2 + \frac{2}{5}a^2b^3\\
&= \frac{3}{5}a^2 + \frac{2}{5}a^2\\
&= a^2\\
\end{split}
\end{equation*}
